CMB plotting

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Planck 2018 data

Obtained from https://cmb.wintherscoming.no/data/, 	COM_PowerSpect_CMB-TT-full_R3.01.txt

```python
#df = pd.read_fwf('plot\COM_PowerSpect_CMB-TT-full.txt', header=None)
#df.to_csv('plot\COM_PowerSpect_CMB-TT-full.csv', index=False, header=['#','l','Dl','-dDl','+dDl'])
```

```python
df  = pd.read_csv("plot\COM_PowerSpect_CMB-TT-full.csv")
df.keys()
```

```python
y_errormin =df['mdDl'][::20]
y_errormax =df['pdDl'][::20]
yerr =[y_errormin, y_errormax]

plt.errorbar(df['l'][::20], df['Dl'][::20], yerr,fmt ='.', capsize=3)
plt.plot(df['l'][::20],df['Dl'][::20],'r-')
```

Binned data, from same source, COM_PowerSpect_CMB-TT-binned_R3.01.txt	

```python
#df = pd.read_fwf('plot\COM_PowerSpect_CMB-TT-binned.txt', header=None)
#df.to_csv('plot\COM_PowerSpect_CMB-TT-binned.csv', index=False, header=['#','l','Dl','mdDl','pdDl','BestFit'])
```

```python
dff  = pd.read_csv("plot\COM_PowerSpect_CMB-TT-binned.csv")
dff.keys()
```

```python
y_errormin =dff['mdDl']
y_errormax =dff['pdDl']
yerr =[y_errormin, y_errormax]

plt.errorbar(dff['l'], dff['Dl'], yerr,fmt ='.', capsize=3)
plt.plot(dff['l'],dff['Dl'],'r-')
```

Planck 2013 data

Obtained from https://irsa.ipac.caltech.edu/data/Planck/release_1/ancillary-data/previews/COM_PowerSpect_CMB_R1.10/index.html

```python
#df = pd.read_fwf('plot\COM_PowerSpect_CMB_R1.10.txt', header=None)
#df.to_csv('plot\COM_PowerSpect_CMB_R1.10.csv', index=False, header=['ell','lmin','lmax','D_ell','err'])
```

```python
dfff  = pd.read_csv("plot\COM_PowerSpect_CMB_R1.10.csv")
dfff.keys()
```

```python
y_errormin =dfff['ell']-dfff['lmin']
y_errormax =dfff['lmax']-dfff['ell']
yerr =[y_errormin, y_errormax]

plt.errorbar(dfff['ell'], dfff['D_ell'], yerr,fmt ='.', capsize=3)
plt.plot(dfff['ell'],dfff['D_ell'],'r-')
```
